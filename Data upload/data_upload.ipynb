```python
import sagemaker
import boto3
```

```python
sm_boto3=boto3.client("sagemaker")
sess=sagemaker.Session()
region=sess.boto_session.region_name
print("Region - "+region)
bucket="tastenet-restaurant-ranker" # Mention the created S3 bucket name here
print("Using bucket - "+bucket)
```

DATASET SOURCE -

KAGGLE

SAVED IN "../Training data" as Restaurant_reviews.csv

```python
import pandas as pd
from tqdm import tqdm

chunksize = 1000
df_list = []

for chunk in tqdm(pd.read_csv("../Training data/Restaurant_reviews.csv", chunksize=chunksize), desc="Reading CSV"):
    df_list.append(chunk)

df = pd.concat(df_list, ignore_index=True)
```
